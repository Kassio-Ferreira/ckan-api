## Testando API do CKAN no Python

* Passo 1: Fazer uma requisição para obter um conjunto de dados:

```python
import requests
import pandas as pd

# fazendo uma requisicao e recebendo um json do ckan:
#r = requests.get('https://demo.ckan.org/api/3/action/datastore_search?resource_id=c655e67b-081a-4de4-a42e-7176be10cd8b&limit=5')
r = requests.get('http://dadosabertos.ufersa.edu.br/api/action/datastore_search?resource_id=92240558-3a69-4a1d-a93a-c448fbb39519')

r = r.json()
dados = r['result']['records'] 
#print(dados)

# convertendo o dicionário recebido num pandas dataframe:
pd.DataFrame.from_dict(dados, orient="columns")
```
